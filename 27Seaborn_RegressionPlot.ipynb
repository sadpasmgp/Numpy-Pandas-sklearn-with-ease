```python
import numpy as np
from numpy.random import randn
import pandas as pd

#stats lib
from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
plt.style.use('ggplot')
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,order=4,
           scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
tips.head()
```

```python
tips['perc']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','perc',tips)
```

```python
sns.lmplot('size','perc',tips,x_jitter=.1)
```

```python
sns.lmplot('size','perc',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','perc',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','perc',tips,hue='day')
```
