```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import webbrowser
webbrowser.open('https://en.wikipedia.org/wiki/List_of_all-time_NFL_win%E2%80%93loss_records')
```

Now on that website copy NFL data to clipboard

```python
nfl_data = pd.read_clipboard()
```

```python
nfl_data
```

```python
nfl_data.index
```

```python
nfl_data.columns
```

We will correct the column names as they have extra spaces

```python
nfl_data.columns = [x.strip() for x in nfl_data.columns]
```

```python
nfl_data.columns
```

Now we will set the Ranks as index for the data

```python
nfl_data.set_index('Rank', inplace=True)
```

```python
nfl_data
```

Now we can access rows using Ranks

```python
nfl_data.loc[7:9]
```

we can also reset index if we like to go back to previous index

```python
nfl_data.reset_index(inplace=True)
```

```python
nfl_data
```

We can reindex as well

```python
Ser = Series(['USA','Mexico','Canada'],index = [0,5,10])
```

```python
Ser
```

```python
ranger = list(range(20))
ranger
```

```python
Ser.reindex(ranger)
```

```python
Ser.reindex(ranger, method = 'ffill') #forwardfill with previous value
```

```python
dframe = DataFrame(np.random.randn(25).reshape(5,5),index=['A','B','C','D','F'],
                  columns=['a','b','c','d','e'])
```

```python
dframe
```

```python
dframe2= dframe.reindex(['A','B','C','D','E','F'])
dframe2
```

As indicated above when reindexing is done, pandas with auto fill with NAN if the index is not found

```python
new_columns = ['a','b','c','col1','col2','col3']
```

```python
dframe2.reindex(columns=new_columns)
```
