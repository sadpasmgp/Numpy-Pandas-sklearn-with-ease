```python
import numpy as np
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
flight = sns.load_dataset('flights')
```

```python
flight.head()
```

```python
flight = flight.pivot('month','year','passengers')
flight
```

```python
sns.heatmap(flight)
```

```python
sns.heatmap(flight,annot=True)
```

```python
sns.heatmap(flight,annot=True,fmt='d')
```

```python
sns.heatmap(flight,center=flight.loc['January',1955],annot=True,fmt='d')
```

```python
f,(axis1,axis2) = plt.subplots(2,1)

yearly_flights = flight.sum()

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

year_dframe = pd.concat((years,flights),axis=1)
year_dframe.columns = ['Year','Flights']

sns.barplot('Year',y='Flights',data=year_dframe,ax=axis1)

sns.heatmap(flight,cmap='Blues',ax=axis2)
```
