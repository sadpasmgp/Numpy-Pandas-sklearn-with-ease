From the data GP - Games played dtype is an object while it should be int64

```python
nfl_data.GP.dtype
```

GP is in strings, we will need to remove extra comma and retain digits

```python
nfl_data.GP = nfl_data.GP.replace(',',"",regex=True)
```

```python
nfl_data.GP
```

Now change the dtype to int64

```python
nfl_data.GP = nfl_data.GP.astype('int64')
```
